```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.metrics.cluster import pair_confusion_matrix
from sklearn import metrics
from sklearn_extra.cluster import KMedoids
import sklearn.metrics.pairwise as smp
import sklearn.cluster as sc
import sentence_transformers as st
import csv
```

```python
sentences = []
truth_labels = []
with open("../comments_2.csv", newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        sentences.append(row[0].lower())
        truth_labels.append(row[1])
n = len(sentences)
```

```python
def cluster_and_evaluate(texts,K,X,truth_labels,p = False):
    clustering = KMedoids(n_clusters=K,method='pam',metric='cosine',init='k-medoids++')
    clustering.fit(X)
    print("\n")
    print("=== Unsupervised Metric ===\n")
    print("== Silhouette Coefficient [0,1*]==")
    print(metrics.silhouette_score(X, clustering.labels_, metric='cosine'),"\n")
    print("\n")
    print("=== Supervised Metric ===\n")
    print("== V-measure [0,1*] ==")
    print(metrics.v_measure_score(truth_labels, clustering.labels_),"\n")
    print("\n")
    print("=== Clusters Exemplars ===\n")
    for i,t in enumerate(clustering.medoid_indices_):
        print(i,": ",texts[t],"\n")
```

```python
print("=== SentenceTransformer ===\n")
model = st.SentenceTransformer('paraphrase-distilroberta-base-v1')
embeddings = model.encode(sentences)
cluster_and_evaluate(sentences,90,embeddings,truth_labels)
```
